```python
import pandas as pd
data=pd.read_csv("https://raw.githubusercontent.com/aniruddhachoudhury/Red-Wine-Quality/master/winequality-red.csv")
df=data
```

```python
data
```

```python
import numpy as np
import seaborn as sns 
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df.info()
```

```python
df.describe()
```

```python
X=df.drop(["quality"],axis=1)
```

```python
Y=df["quality"]
```

```python
X.head()
```

```python
Y.head()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.33,random_state=42)
```

```python
x_test.head()
```

```python
x_train.head()
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit(x_train)
```

```python
scaler.mean_
```

```python
scaler.transform(x_train)
```

```python
#or else we can directly use fit tranform method
x_train_tf=scaler.fit_transform(x_train)
```

```python
x_train_tf
```

```python
from sklearn.svm import SVC
model=SVC()
model.fit(x_train_tf,y_train)
```

```python
model.score(x_train_tf,y_train)
```

```python
x_test_tf=scaler.transform(x_test)
```

```python
y_pred=model.predict(x_test_tf)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,y_pred)
```

## here we are getting onluy 59% precuuracy so let us chech with the logistic regression whether the accuracy will change with respect to the logistic regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
model2=LogisticRegression()
```

```python
model2.fit(x_train_tf,y_train)
```

```python
y_pred_2=model2.predict(x_test_tf)
```

```python

accuracy_score(y_test,y_pred_2)
```
